```python
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
```

```python
#Read in FLUXNET output from 12/03/21
os.chdir('/Users/graciepearsall/Elkhorn-Slough-EC/YMA/eddy-pro-files/YMA_20211203')
yma_1203 = pd.read_csv('eddypro_out_fluxnet_2021-12-07T150428_exp.csv')
yma_1203.head()
```

```python
#Read in FLUXNET output from 12/19/21
os.chdir('/Users/graciepearsall/Elkhorn-Slough-EC/YMA/eddy-pro-files/YMA_20211219')
yma_1219 = pd.read_csv('eddypro_20211219_output_fluxnet_2022-04-04T150916_exp.csv')
yma_1219.tail()
```

```python
#Read in FLUXNET output from 1/11/22
os.chdir('/Users/graciepearsall/Elkhorn-Slough-EC/YMA/eddy-pro-files/YMA_20220111')
yma_0111 = pd.read_csv('eddypro_yma_221219_fluxnet_2022-03-30T102236_exp.csv')
yma_0111.tail()
```

```python
#Combine Frames
frames = [yma_1203, yma_1219, yma_0111]
yma_full = pd.concat(frames)
yma_full.drop
```

```python
#Add datetime object and replace null values
yma_full['DATE'] = pd.to_datetime(yma_full.TIMESTAMP_END, format= '%Y%m%d%H%M')
yma_full.replace(-9999.0, np.nan, inplace=True)
```

```python
os.chdir('/Users/graciepearsall/Elkhorn-Slough-EC/YMA/notebooks/YMA_synthesis')
```

```python
#Plot FCH4 and CH4
fig, ax= plt.subplots(2, figsize=(15,15))
ax1 = ax[0]
ax2 = ax[1]

ax1.plot(yma_full.DATE, yma_full.FCH4)
ax1.set_ylabel('nmol CH4 m-2 30 min-1')
ax1.set_title('FCH4', fontsize=16)
plt.grid()

ax2.plot(yma_full.DATE, yma_full.CH4, color='tab:orange')
ax2.set_title('CH4', fontsize=16)
ax2.set_ylabel('nmol CH4 mol-1')
plt.grid()
plt.savefig('CH4_YMA_Full.png', dpi=400)
```

```python
#Plot CO2 
fig, ax= plt.subplots(2, figsize=(15,15))
ax1 = ax[0]
ax2 = ax[1]

ax1.plot(yma_full.DATE, yma_full.FC, color='darkgreen')
ax1.set_ylabel('µmolCO2 m-2 30min-1')
ax1.set_title('FC', fontsize=16)
plt.grid()

ax2.plot(yma_full.DATE, yma_full.CO2, color='tab:red')
ax2.set_title('CO2', fontsize=16)
ax2.set_ylabel('nmolCO mol-1')
plt.grid()
plt.savefig('CO2_YMA_Full.png', dpi=400)
```

```python
#Plot H2O 
fig, ax= plt.subplots(2, figsize=(15,15))
ax1 = ax[0]
ax2 = ax[1]

ax1.plot(yma_full.DATE, yma_full.FH2O, color='tab:purple')
ax1.set_ylabel('µmolCO2 m-2 30min-1')
ax1.set_title('FH20', fontsize=16)
plt.grid()

ax2.plot(yma_full.DATE, yma_full.CO2, color='darkblue')
ax2.set_title('H2O', fontsize=16)
ax2.set_ylabel('mmolH2O mol-1 ')

plt.grid()
plt.savefig('H2O_YMA_Full.png', dpi=400)
```
